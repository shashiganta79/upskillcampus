```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Load the clean dataset
data_dir = '/kaggle/input/datasets/shashiganta/crop-production-data/'
df = pd.read_csv(data_dir + 'datafile (1).csv')

# Clean up any accidental trailing spaces in column names
df.columns = df.columns.str.strip()

# 2. Separate our Features (X) and Target Variable (y)
# We will predict 'Yield (Quintal/ Hectare)'
X = df[['Crop', 'State', 'Cost of Cultivation (`/Hectare) A2+FL', 
        'Cost of Cultivation (`/Hectare) C2', 'Cost of Production (`/Quintal) C2']]
y = df['Yield (Quintal/ Hectare)']

# 3. Preprocess Text Data (One-Hot Encoding for 'Crop' and 'State')
categorical_features = ['Crop', 'State']
numerical_features = ['Cost of Cultivation (`/Hectare) A2+FL', 
                      'Cost of Cultivation (`/Hectare) C2', 
                      'Cost of Production (`/Quintal) C2']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ], remainder='passthrough' # Keep numerical features as they are
)

# Apply transformation to our features
X_processed = preprocessor.fit_transform(X)

# 4. Split data into Training set (80%) and Testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# 5. Train the Machine Learning Model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 6. Evaluate Model Performance
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("--- MODEL TRAINING COMPLETED ---")
print(f"Mean Absolute Error (MAE): {mae:.2f} Quintals/Hectare")
print(f"R-squared (Accuracy) Score: {r2 * 100:.2f}%")
```
